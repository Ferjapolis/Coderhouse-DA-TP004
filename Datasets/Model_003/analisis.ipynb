```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
def detectar_estado_animo(cancion):
    try:
        sid = SentimentIntensityAnalyzer()
        score = sid.polarity_scores(cancion)
        if score['compound'] >= 0.05:
            return 'feliz'
        elif score['compound'] > -0.05 and score['compound'] < 0.05:
            return 'neutro'
        else:
            return 'triste'
    except:
        print(cancion)
```

```python
endsong['estado_animo'] = endsong['master_metadata_track_name'].apply(detectar_estado_animo)
```

```python
endsong['fecha'] = pd.to_datetime(endsong['ts']).dt.date
resumen = endsong.groupby(['fecha', 'estado_animo'])['min'].sum().reset_index()
```

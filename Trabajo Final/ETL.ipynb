```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
usuarios = pd.read_excel(r"MOCK_DATA.xlsx",engine='openpyxl',sheet_name="Usuarios")
print("Usuarios:",usuarios.shape)
print(list(usuarios.columns.values))

registros = pd.read_excel(r"MOCK_DATA.xlsx",engine='openpyxl',sheet_name="Registros")
print("Registros:",registros.shape)
print(list(registros.columns.values))
```

```python
usuarios = usuarios.rename(columns={"CIDW":"nominador"})
data = pd.merge(registros, usuarios, on="nominador", how="left")
columnas = {'Nombre': 'NR_name', 
            'Apellido': 'NR_name',
            'Externo': 'NR_last',
            'Genero': 'NR_gen',
            'Area': 'NR_area',
            'Email': 'NR_mail',
            'Contracto': 'NR_Contract'}
data = data.rename(columns=columnas)
```

```python
usuarios = usuarios.rename(columns={"nominador":"nomidado"})
data = pd.merge(data, usuarios, on="nomidado", how="left")
columnas = {"nomidado":"nominado",
            'Nombre': 'NN_name', 
            'Apellido': 'NN_name',
            'Externo': 'NN_last',
            'Genero': 'NN_gen',
            'Area': 'NN_area',
            'Email': 'NN_mail',
            'Contracto': 'NR_Contract'}
data = data.rename(columns=columnas)
```

```python
data.head()
```

```python
data['periodo']=data['fecha'].apply(lambda x: x.strftime('%Y-%m'))
grupo_periodo = data.groupby(['periodo'])
resumen = grupo_periodo["NR_area"].value_counts()
```
